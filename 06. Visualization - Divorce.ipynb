Import Libraries for Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Question 6
Create a line plot showing the number of marriages and divorces per capita in the 
U.S. between 1867 and 2014. Label both lines and show the legend.
Don't forget to label your axes!

```python
# Import file
divorce_data = pd.read_csv('us-marriages-divorces-1867-2014.csv')

years = divorce_data['Year'].values
marriages = divorce_data['Marriages_per_1000'].values
divorces = divorce_data['Divorces_per_1000'].values

# Fill null values as '0'
divorce_data.fillna(0, inplace=True)
```

```python
# Set which columns will be viewed
dataset = divorce_data[['Year', 'Marriages_per_1000', 'Divorces_per_1000']]

dataset
```

```python
# Get columns to plot
columns = dataset.columns.drop(['Year'])

# Create x data
x_data = x_data = dataset['Year']

# Create figure and axis
fig, ax = plt.subplots()

# Plot each column
for column in columns:
    ax.plot(x_data, dataset[column], label=column)

# Set title and legend
ax.set_title('Marriages and Divorces per Capita in the U.S.')
ax.set_xlabel('Year')
ax.set_ylabel('Per Capita Rate')
ax.legend()
plt.grid(True)


# display the plot
plt.show()
```

Question 7
Create a vertical bar chart comparing the number of marriages and divorces per 
capita in the U.S. between 1900, 1950, and 2000.
Don't forget to label your axes

```python
comp_data = divorce_data[
    divorce_data['Year'].apply(lambda x: x in [1900, 1950, 2000])]

comp_dataset = comp_data[['Year', 'Marriages_per_1000', 'Divorces_per_1000']]
```

```python
print(comp_dataset)
```

```python
# Set up the figure and axis
fig, ax = plt.subplots()

# Create vertical bar chart
bar_width = 10
bar_positions_marriages = comp_dataset['Year']
bar_positions_divorces = [pos + bar_width for pos in bar_positions_marriages]

ax.bar(bar_positions_marriages, comp_dataset['Marriages_per_1000'], width=bar_width, label='Marriages')
ax.bar(bar_positions_divorces, comp_dataset['Divorces_per_1000'], width=bar_width, label='Divorces')

# Set labels and title
ax.set_xlabel('Year')
ax.set_ylabel('Per Capita Rate')
ax.set_title('Marriages and Divorces per Capita in the U.S. (1900, 1950, 2000)')
ax.set_xticks([pos + bar_width / 2 for pos in bar_positions_marriages])
ax.set_xticklabels(comp_dataset['Year'])
ax.legend()

# Display the plot
plt.grid(True)
plt.show()
```
