Question 8
Create a horizontal bar chart that compares the deadliest actors in Hollywood. Sort 
the actors by their kill count and label each bar with the corresponding actor's name.
Don't forget to label your axes!

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import file
kill_count = pd.read_csv('actor_kill_counts.csv')

kill_count
```

```python
# Sort the DataFrame by 'Count' column
kill_count_sorted = kill_count.sort_values(by='Count', ascending=True)

# Create a horizontal bar chart
plt.barh(kill_count_sorted['Actor'], kill_count_sorted['Count'], color='red')

# Label axes and title
plt.xlabel('Kill Count')
plt.ylabel('Actor')
plt.title('Deadliest Actors in Hollywood')

# Display the plot
plt.show()
```

Question 9
Create a pie chart showing the fraction of all Roman Emperors that were assassinated.
Make sure that the pie chart is an even circle, labels the categories, and shows the percentage breakdown of the categories.

```python
# Import File
emperor = pd.read_csv('roman-emperor-reigns.csv')

emperor
```

```python
# Count the occurrences of each cause of death
cause_count = emperor['Cause_of_Death'].value_counts().sort_values(ascending=True)

# Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(cause_count,
        labels=cause_count.index,
        autopct='%1.1f%%',
        startangle=90,
        )

# Title
plt.title('Roman Emperor Cause of Death')

# Display the pie chart
plt.show()
```

Question 10
Create a scatter plot showing the relationship between the total revenue earned by 
arcades and the number of Computer Science PhDs awarded in the U.S. between 
2000 and 2009.
Don't forget to label your axes! Color each dot according to its year

```python
# Import File
arcade = pd.read_csv('arcade-revenue-vs-cs-doctorates.csv')

arcade
```

```python
# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x = 'Total Arcade Revenue (billions)',
                y = 'Computer Science Doctorates Awarded (US)',
                data = arcade,
                hue = 'Year',
                size='Year',
                sizes=(200, 201),
                palette = 'coolwarm')

# Title and labels
plt.title('Relationship Between Arcade Revenue and Computer Science Doctorates Awarded (2000-2009)')
plt.xlabel('Total Arcade Revenue (billions)')
plt.ylabel('Number of Computer Science Doctorates Awarded (US)')

# Display the plot
plt.show()
```
